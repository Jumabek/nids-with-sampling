```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from os.path import join
%matplotlib inline
```

```python
def matplotlib_setup():
    figure(figsize=(18,6))
    font = {'size'   : 35}
    plt.rc('font', **font)
    plt.rcParams["axes.linewidth"]  = 2.5
    plt.ylim([50,120])
    plt.grid(linewidth=3,axis='y', color='grey')
    
    CB91_Blue = '#2CBDFE'
    CB91_Green = '#47DBCD'
    CB91_Pink = '#F3A0F2'
    CB91_Purple = '#9D2EC5'
    CB91_Violet = '#661D98'
    CB91_Amber = '#F5B14C'
    color_list = [CB91_Blue, CB91_Pink, CB91_Green, CB91_Amber,
              CB91_Purple, CB91_Violet]
    plt.rcParams['axes.prop_cycle'] = plt.cycler(color=color_list)
    plt.yticks(np.arange(50,101,10),np.arange(50,101,10) )
dataroot = '/data/juma/data/ids18/CSVs_r_1.0_m_1.0'
patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]
```

```python
def plot_bars(dt_bars, rf_bars, cnn_bars):
    
    matplotlib_setup()

    # set width of bar
    barWidth = 0.25

    # Set position of bar on X axis
    r1 = np.arange(len( dt_bars))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
 
    # Make the plot
    plt.bar(r1, dt_bars,  width=barWidth, edgecolor='white', label='Decision Tree', hatch=patterns[0])
    plt.bar(r2,  rf_bars, width=barWidth, edgecolor='white', label='Random Forest',hatch=patterns[1])
    plt.bar(r3,  cnn_bars,  width=barWidth, edgecolor='white', label='CNN',hatch=patterns[2])
 
    # Add xticks on the middle of the group bars
    plt.xticks([r + barWidth for r in range(len( dt_bars))], ['NS', 'SRS', 'FFS', 'SFS', 'SGS'])
    
    # Create legend & Show graphic
    plt.legend(ncol=3,fancybox=True, framealpha=0.3)
    plt.tight_layout()
    return plt
```

```python
# ANY TPR
dt_bars = [99.09, 70.04, 75.48, 84.3, 83.54]
rf_bars = [99.09, 82.03, 80.69, 88.3, 87.02]
cnn_bars = [96.74, 75.83, 78.1, 86.38, 82.38]

plot_bars(dt_bars, rf_bars, cnn_bars)

plt.ylabel('Detection Rate (%)\n Flow Det. Met.: Any')
plt.savefig(join(dataroot,'tpr_any.png'))
plt.show()
```
