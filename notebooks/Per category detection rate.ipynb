```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from os.path import join
%matplotlib inline
```

```python
def matplotlib_setup():
    figure(figsize=(60,12))
    font = {'size'   : 45}
    plt.rc('font', **font)
    plt.rcParams["axes.linewidth"]  = 2.5
    plt.ylim([0,120])
    plt.grid(linewidth=4,axis='y', color='grey')
    
    CB91_Blue = '#2CBDFE'
    CB91_Green = '#47DBCD'
    CB91_Pink = '#F3A0F2'
    CB91_Purple = '#9D2EC5'
    CB91_Violet = '#661D98'
    CB91_Amber = '#F5B14C'
    color_list = [CB91_Blue, CB91_Pink, CB91_Green, CB91_Amber,
              CB91_Purple, CB91_Violet]
    plt.rcParams['axes.prop_cycle'] = plt.cycler(color=color_list)

dataroot = '/data/juma/data/ids18/CSVs_r_1.0_m_1.0'
patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]
```

```python
matplotlib_setup()
#ALL
dataroot = '/data/juma/data/ids18/'
# set width of bar
barWidth = 0.18
 
# 10%
srs_bars = [53.58,99.84,81.23,85.75,79.56,94.72,67.4,30.8,60,73.08,94.02,99.78,99.89,0,74.31]
sfs_bars = [58.87,99.76,95.06,97.23,94.43,99.82,75.95,33.03,70,88.46,94.02,99.71,99.46,0,80.53]
ffs_bars = [54.67,99.73,78.27,86.75,82.96,95.71,71.19,31.63,30,73.08,85.47,99.67,99.82,0,71.87]
sgs_bars = [60.79,98.17,93.8,98.26,95.78,97.12,86.99,31.08,40,94.23,81.2,99.67,99.85,0,78.17]



# Set position of bar on X axis
r1 = np.arange(len( srs_bars))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
 
# Make the plot
plt.bar(r1, srs_bars,  width=barWidth, edgecolor='white', label='Simple Random Sampling (SRS)', hatch=patterns[0])
plt.bar(r2,  sfs_bars, width=barWidth, edgecolor='white', label='SketchFlow Sampling (SFS)', hatch=patterns[1])
plt.bar(r3,  ffs_bars, width=barWidth, edgecolor='white', label='Fast Filtered Sampling (FFS)', hatch=patterns[2])
plt.bar(r4,  sgs_bars, width=barWidth, edgecolor='white', label='Sketch Guided Sampling (SGS)', hatch=patterns[3])

 
# Add xticks on the middle of the group bars
plt.ylabel('Flow Observation Rate (%)')
#plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len( srs_bars))] ,['Benign','DDoS HOIC','DDoS LOIC HTTP','DoS Golden Eye',
                                                           'DoS Slow HTTP Test','FTP Brute Force','DoS Slowloris',
                                                           'DDoS LOIC UDP','SQL Injection',
                                                           'Brute Force Web','Brute Force XSS','SSH Brute Force',
                                                           'DoS Hulk','Infiltration','Avg. of Malicious ' ],rotation=30)
 
# Create legend & Show graphic
plt.legend(ncol=4,fancybox=True, framealpha=0.3)
plt.tight_layout()
plt.savefig(join(dataroot,'det_rate_cat_si_10.png'))
plt.show()
```
