### Goal: compare sampler methods performance
1. We read all the evaluation files
2. extract avg recall for malicous classes
3. make a plot for each sr x classifier 


```python
import pandas as pd
import numpy as np
import glob 
from os.path import join
import os
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_immediate_subdirs(a_dir, only=''):
    if only=='':
        return [os.path.join(a_dir, name) for name in os.listdir(a_dir) 
                if os.path.isdir(os.path.join(a_dir, name)) and name.endswith('_l') ]
    else:
        return [os.path.join(a_dir, name) for name in os.listdir(a_dir)
                if os.path.isdir(os.path.join(a_dir, name)) and name.endswith('_l') and only in name]

def ensure_dir(dir_path):
    if not os.path.exists(dir_path):
        os.makedirs(dir_path)
```

```python
def get_pkt_count(dirs,ending):
    counts = []
    for d in dirs:
        num_sampled_pkts = 0
        for f in glob.glob(join(d,ending)):
            num_sampled_pkts += int(open(f).readline())
        counts.append(num_sampled_pkts)
    return counts
```

```python
def get_accuracy(dirs,classifier_name):
    accuracies = []
    for d in dirs:
        df = pd.read_csv(join(d,classifier_name,'evaluation.csv'))
        df_avg = df.tail(1)
        acc = df_avg['Rc'].values[0]
        accuracies.append(acc)
    return accuracies
    
```

```python
def get_acc_dir_names(dirs,ending,classifier_name):
    pkt_counts = np.array(get_pkt_count(dirs,ending))
    accuracies = np.array(get_accuracy(dirs,classifier_name))
    dir_names = np.array([os.path.basename(dir)[:-2] for dir in dirs])
        
    sorted_index = np.argsort(pkt_counts)
    pkt_counts = pkt_counts[sorted_index]
    accuracies = accuracies[sorted_index]

    dir_names = dir_names[sorted_index]
    return accuracies,pkt_counts,dir_names
```

```python

def autolabel(ax,rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0, 'right': 1, 'left': -1}
    max_height = 0
    for rect in rects:
        height = rect.get_height()
        if height > max_height:
            max_height = height
            
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{0:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(offset[xpos]*3, 1),  # use 3 points offset
                    textcoords="offset points",  # in both directions
                    ha=ha[xpos], va='bottom')
```

```python

def plot_bar(filename,ind,counts,colors,dir_names,label,width,legend,figsize=(10,5)):
    plt.clf()
    plt.close()
    fig, ax = plt.subplots(nrows=1,ncols=1,figsize=figsize)
    sorted_index = np.argsort(dir_names)
    counts = counts[sorted_index]
    colors = colors[sorted_index]
    dir_names = dir_names[sorted_index]
    
    rects = ax.bar(ind,counts,color=colors,label=legend,width=width)
    ax.set_xticks(ind)
    ax.set_xticklabels(dir_names,rotation=45)
    #ax.set_xlabel(label)
    ax.legend()
    ax.set_ylabel('avg recall - balanced accuracy')
    autolabel(ax,rects)
    ax.margins(0.15)
    ax.set_ylim(0,1)
    plt.tight_layout()
    plt.grid()
    plt.savefig(filename,dpi=100)
    plt.show()
    
    df = pd.DataFrame.from_dict({'sampler':dir_names,'avg-Recall':counts})
    df = df[['sampler','avg-Recall']]
    df.to_csv(filename.replace('.png','.csv'),index=False)

    
```
