```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from os.path import join
%matplotlib inline
```

```python
def matplotlib_setup():
    figure(figsize=(50,10))
    font = {'size'   : 35}
    plt.rc('font', **font)
    plt.rcParams["axes.linewidth"]  = 2.5
    plt.ylim([0,120])
    plt.grid(linewidth=3,axis='y', color='grey')
```

```python
matplotlib_setup()

dataroot = '/data/juma/data/ids18/'
# set width of bar
barWidth = 0.18
 
# 10%
srs_bars = [53.68,99.79,81.23,85.75,79.56,94.72,67.4,30.8,60,99.78,99.89,0,73.08,94.02,74.31]
sfs_bars = [58.91,99.75,95.06,97.23,94.43,99.82,75.95,33.03,70,99.71,99.46,0,88.46,94.02,80.53]
ffs_bars = [54.54,99.87,78.28,86.75,82.96,95.71,71.12,31.63,40,99.78,99.85,0,76.92,81.2,72.62]
sgs_bars = [60.74,98.16,93.8,98.26,95.78,97.12,87.06,31.08,30,99.67,99.85,0,94.23,81.2,77.4]


# Set position of bar on X axis
r1 = np.arange(len( srs_bars))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
 
# Make the plot
plt.bar(r1, srs_bars, color='#7f6d5f', width=barWidth, edgecolor='white', label='SRS')
plt.bar(r2,  sfs_bars, color='#557f2d', width=barWidth, edgecolor='white', label='SFS')
plt.bar(r3,  ffs_bars, color='#2d7f5e', width=barWidth, edgecolor='white', label='FFS')
plt.bar(r4,  sgs_bars, color='#5c7f5e', width=barWidth, edgecolor='white', label='SGS')
 
# Add xticks on the middle of the group bars
plt.ylabel('Flow Observation Rate (%)')
#plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len( srs_bars))] ,['Benign','DDoS HOIC','DDoS LOIC HTTP','DoS Golden Eye',
                                                           'DoS Slow HTTP Test','FTP Brute Force','DoS Slowloris',
                                                           'DDoS LOIC UDP','SQL Injection',
                                                           'Brute Force Web','Brute Force XSS','SSH Brute Force',
                                                           'DoS Hulk','Infiltration','Avg. of Malicious ' ],rotation=30)
 
# Create legend & Show graphic
plt.legend(ncol=4,fancybox=True, framealpha=0.3)
plt.tight_layout()
plt.savefig(join(dataroot,'det_rate_cat_si_10.png'))
plt.show()
```
